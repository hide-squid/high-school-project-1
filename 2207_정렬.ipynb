```python
arr = [1,3,4,5,2] #arr를 리스트로 정의
def selection_sort(arr): #함수 selection_sort를 arr와 함께 정의
    for i in range(len(arr)-1): #(len(arr)-1)번 반복
        min_idx = i #min_idx를 i로 정의
        for j in range(i+1,len(arr)): #i+1에서부터 4까지 반복
            if arr[j] < arr[min_idx]: #arr[j]가 arr[min_idx]보다 작을때
                min_idx = j #min_idx를 j로 정의
        arr[i],arr[min_idx] = arr[min_idx],arr[i] #arr[i] arr[min_idx] 위치 바꾸기
    print(arr) #arr배열 출력
selection_sort(arr) #selection_sort 호출
```

```python
array = [1,3,4,5,2]
def bubble_sort(arr):
    for k in range(len(arr)-1, 0, -1):
        for m in range(k):
            if arr[m]>arr[k]:
                arr[m],arr[k] = arr[k],arr[m]
    print(arr)
bubble_sort(array)
```
