```python
#계산기
num1 = int(input('숫자1 입력: '))
num2 = int(input('숫자2 입력: '))

sum = num1 + num2
min = num1 - num2
mul = num1 * num2
div1 = num1 / num2
div2 = num1 // num2  #몫
div3 = num1 % num2   #나머지

print("{} + {} = {}".format(num1, num2, sum))
print("{} - {} = {}".format(num1, num2, min))
print("{} * {} = {}".format(num1, num2, mul))
print("{} / {} = {}".format(num1, num2, div1))
print("{} // {} = {}".format(num1, num2, div2))
print("{} % {} = {}".format(num1, num2, div3))
```

```python
class Student():
    def __init__(self, name, age, like):
        self.name = name
        self.age = age
        self.like = like
    def studentInfo(self):
        print(f'이름:{self.name}, 나이:{self.age}, 좋아하는 것:{self.like}')
김철수 = Student('김철수', 17, '축구')
장다인= Student('장다인', 5, '헬로카봇')
김철수.studentInfo()
장다인.studentInfo()
```

```python
class Mother():
    def charateristic(self):
        print('키가 크다')
        print('공부를 잘한다.')
class Daughter(Mother):
    def charateristic(self):
        super().charateristic()
        print('운동을 잘한다.')
엄마 = Mother()
딸 = Daughter()
print('엄마는')
엄마.charateristic()
print('딸은')
딸.charateristic()
```

```python
import random
r_number = random.randint(1, 100)
cnt = 1
while True:
    try:
        my_number = int(input('숫자입력:'))
        if my_number>r_number:
            print('down')
        elif my_number<r_number:
            print('up')
        elif my_number == r_number:
            print(f"정답!, {cnt}회 만에 맞췄습니다.")
            break
        cnt += 1
    except:
        print('입력오류, 숫자를 입력하세요!')
```
