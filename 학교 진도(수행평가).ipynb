```python
# [1번] 1부터 20까지 합산하는 프로그램을 작성하시오.
# 변수명: 자유롭게 사용
# while문으로 작성할 것
i = 0             #종료알려주는 변수
sum = 0           #전체 합
while True:       #무한반복
    sum += i      #전체 더한것에 i더하기
    i+=1          #i에 1씩 더하기ㅣ
    if i > 20:    #만약 i가 20이면
        break     #종료
print(sum)        #전체 더한거 출력
```

```python
# [2번] 출생년도를 입력받아서 나이를 출력하고 성인(20세이상), 청소년(13세이상), 어린이(13세미만)를 구분하여 출력하시오.
# 변수명: 자유롭게 사용

year = int(input("출생년도를 입력하세요: ")) #year이라는 변수에 정수로 년도받기
this_year = 2023                        #this_year에 2023넣기
year1 = int(this_year) - int(year)      #2023이랑 받은 년도 빼기
print("%d살" %year1)                     #몇살인지 출력하기
if year1>=20:                           #만약 나이가 20이상이면
    print("성인")                        #성인 출력
elif year1>=13:                         #만약 나이가 13이상이며
    print("청소년")                      #청소년 출력
else:                                   #만약 나이가 13미만이라며
    print("어린이")                      #어린이 출력
```

```python
# [3번] 2단부터 9단까지 구구단을 출력하는 프로그램을 작성하시오.
# 변수명: 자유롭게 사용
# 단별로 세로로 출력하시오.

for a in range(1, 10):                          #9번 반복시켜라
    for b in range(2, 10):                      #8번 반복시켜라
        print("%dX%d=%d" %(b, a, a*b), end='\t')#b가 곱하기 앞에 숫자 / a가 곱하기 뒤에 숫자
                                                #end='\t'는 뒤에 네칸을 띄어라
    print()                                     #다음 줄로 내려

#여기서는 숫자랑 기호 바꿔봐야함!!!
```

```python
# [4번] 단수를 입력받아서 거꾸로 별삼각형 출력하기
# 변수명: 자유롭게 사용
dan = int(input("가로: "))
for i in range(dan, 0, -1):
    for j in range(0, i):
        print("*",end ='')
    print()
```
