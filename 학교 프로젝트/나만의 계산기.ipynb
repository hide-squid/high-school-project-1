이름하여 **천척산(天尺算)** 이노라


**1)** 연주시차를 이용하여 별이나 행성이 지구에게서 얼마나 떨어져있는지 구할수 있는 계산기 제작

**2)** 중2, 3? 과학시간에 별에 대해 배웠었는데 신기하고 궁금해서 만들어 보고 싶었던 계산기였다. 그리고 이를 통해 내가 실제로 저 별의 거리를 재어 보고 싶다고 생각 했었을 때가 있었다.

**3)** 함수를 이용해 단위를 끊고 단위를 붙여 받아왔다. 그리고 사용자가 끝내고 싶을때까지 무한반복을 하였고 인터넷에서 연주시차를 파섹, 광년, km로 바꾸는 식을 참고하여 만들었다. 그리고 웬만하면 다 실수이기 때문에 int()가 아닌 float()으로 받았다.

**4)** 이를 통해 연주시차만 알게된다면 쉽게 그 행성까지의 거리를 알수 있다고 판단하고 km로 변환하는 과정에서 10의 거듭제곱으로 인해 숫자가 커져 버린다. 그로 인해 계산이 힘들어지므로 컴퓨터로 만들어 정확도를 올리고 번거러운 일을 최소화 시킬 수 있을 것이다.

```python
def sign(km):
    result = ""
    num = km
    if num >= 1000000000000000000000000: # 1자 이상
        result += ("%d자 " %int(num // 1000000000000000000000000))
        num = num % 1000000000000000000000000

    if num >= 100000000000000000000: # 1해 이상
        result += ("%d해 " %int(num // 100000000000000000000))
        num = num % 100000000000000000000

    if num >= 10000000000000000: # 1경 이상
        result += ("%d경 " %int(num // 10000000000000000))
        num = num % 10000000000000000

    if num >= 1000000000000: # 1조 이상
        result += ("%d조 " %int(num // 1000000000000))
        num = num % 1000000000000

    if num >= 100000000: # 1억 이상
        result += ("%d억 " %int(num // 100000000))
        num = num % 100000000

    if num >= 10000: # 1만 이상
        result += ("%d만 " %int(num // 10000))
        num = num % 10000

    if num >= 1: # 1 이상
        result += ("%d " %int(num // 1))

    return result + 'km'

while 1:
    a = float(input('연주시차 입력("): '))  #2"
    b = int(input("(1. 파섹구하기 2. 광년구하기 3. km구하기 4.끝내기): "))
    if b == 1 or b == 2 or b == 3 or b == 4:
        if a == 0:
            pc = 0
            ly = 0
            km = 0
            if b == 1:
                print("연주시차 -> 파섹(pc) : 약 %fpc" %pc)
            elif b == 2:
                print("연주시차 -> 광년(ly) : 약 %fly" %ly)
            elif b == 3:
                result = sign(km)
                print("연주시차 -> 키로미터(km) : 약 0"+ result)
            elif b == 4:
                break
            print("")

        else:
            pc = 1/a #0.5pc
            ly = pc * 3.2616 #1.63078광년
            km = ly*9460730472580.8 #1.542837e+13 : 15,428,370,000,000km

            if b == 1:
                print("연주시차 -> 파섹(pc) : 약 %fpc" %pc)
            elif b == 2:
                print("연주시차 -> 광년(ly) : 약 %fly" %ly)
            elif b == 3:
                result = sign(km)
                print("연주시차 -> 키로미터(km) : 약 "+ result)
            elif b == 4:
                break
            print("")
    else:
        print("다시 입력")
        print("")
        continue
```
