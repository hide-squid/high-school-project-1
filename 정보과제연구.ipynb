```python
import random
from time import sleep

# 파일 읽기
stu = open('/content/stulist.txt', 'r')
stub = open('/content/boy.txt', 'r')
stug = open('/content/girl.txt', 'r')
output = open('finallist.txt', 'w')

studentblist = []
studentglist = []
teams = [[] for _ in range(6)]  # 6개 팀 준비

# 남학생 리스트
for x in stub:
    a, b = x.split('.')
    studentblist.append(b.strip())

# 여학생 리스트
for x in stug:
    a, b = x.split('.')
    studentglist.append(b.strip())

# 랜덤 시드 입력
seedinput = 0
x = 0
while True:
    x += 1
    try:
        k = int(input('%2d:' % x))
    except ValueError:
        print('only number!')
        x -= 1
    else:
        if k == -1:
            break
        else:
            seedinput += k

random.seed(seedinput)
random.shuffle(studentblist)
random.shuffle(studentglist)

# 2명을 넣을 조 3개 무작위 선택
girl_team_indices = random.sample(range(6), 3)

# 여학생 배치
girl_index = 0
for idx in girl_team_indices:
    teams[idx].extend(studentglist[girl_index:girl_index+2])
    girl_index += 2

# 남학생 배치
boy_index = 0
for i in range(6):
    while len(teams[i]) < 4:
        teams[i].append(studentblist[boy_index])
        boy_index += 1

# 팀 출력
print("\n===== 조 편성 결과 =====\n")
for idx, team in enumerate(teams):
    sleep(0.5)
    print(f"{idx + 1}조: {', '.join(team)}")
    output.write(f"{idx + 1}조: {', '.join(team)}\n")

output.close()
```

```python
from itertools import product

WIN = 5
DRAW = 3
LOSE = 1

games = 3

results = list(product([WIN, DRAW, LOSE], repeat= games))

threshold = 12

passing_cases = [result for result in results if sum(result)>=threshold]

print(f"우리 반이 예선을 통과하는 경우의 수: {len(passing_cases)}")
```

```python
import heapq

def dijkstra(graph, start):
    # 최단 거리 저장 딕셔너리 (무한대 값으로 초기화)
    distances = {node: float('inf') for node in graph}
    distances[start] = 0  # 시작 노드는 거리가 0
    priority_queue = [(0, start)]  # 우선순위 큐 (거리, 노드)

    while priority_queue:
        current_distance, current_node = heapq.heappop(priority_queue)

        # 이미 처리된 노드라면 무시
        if current_distance > distances[current_node]:
            continue

        # 인접 노드 확인
        for neighbor, weight in graph[current_node].items():
            distance = current_distance + weight

            # 더 짧은 경로 발견 시 업데이트
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                heapq.heappush(priority_queue, (distance, neighbor))

    return distances

# 그래프 예제
graph = {
    'A': {'B': 3, 'C': 1},
    'B': {'A': 3, 'C': 4, 'D': 5},
    'C': {'A': 1, 'B': 4},
    'D': {'B': 5}
}

# A 노드로부터 각 노드의 최단 거리 계산
start_node = 'A'
shortest_distances = dijkstra(graph, start_node)
print(f"{start_node}에서 각 노드로의 최단 거리: {shortest_distances}")
```
