```python
def add_record(records):
    record = []
    name = input("이름을 입력하세요: ")
    time = input("지각 시간을 입력하세요 (예: 8:30): ")
    time_ = time.split(":")
    if int(time_[0]) >  7:
        if int(time_[1]) > 25:
            record.append(name)
            record.append(time)
            records.append(record)
            print("저장되었습니다.")
        else:
            return print("지각이 맞는지 확인해주세요!!")
    else:
            return print("지각이 맞는지 확인해주세요!!")
def display_records(records):
    if not records:
        print("기록이 없습니다.")
        return
    print("지각 기록: ")
    for a in records:
        print(a[0] + " - " + a[1])

def count_lates(records):
    print(f"총 지각 횟수: {len(records)}회")

def money_sight(records):
    abc = 0
    index = 0
    money = 0
    name_ = input("지각자 이름을 입력하세요: ")
    for abc in records:
        if name_ in abc:
            index += 1
            money = money + 1000
    print(name_ + "님의 지각비는 %d원입니다." %money)

records = []
while True:
    print("")
    print("지각 관리 프로그램")
    print("1. 지각 기록 추가")
    print("2. 지각 기록 보기")
    print("3. 총 지각 횟수 보기")
    print("4. 지각비 보기")
    print("5. 종료")

    choice = input("\n원하는 기능의 번호를 입력하세요: ")
    print("")

    if choice == '1':
        add_record(records)
    elif choice == '2':
        display_records(records)
    elif choice == '3':
        count_lates(records)
    elif choice == '4':
        money_sight(records)

    elif choice == '5':
        print("프로그램을 종료합니다.")
        break
    else:
        print("잘못된 입력입니다. 다시 시도하세요.")
```
