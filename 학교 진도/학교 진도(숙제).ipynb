```python
#(1)
def binary_search(a,x,cnt): #binary_search 함수 정의
    start = 0 # start에 0 저장
    end = len(a)-1 #end에 len(a)의 길이 저장
    while start <= end: #만약 start가 end보다 작거나 같을때 까지 반복
        mid = (start + end)//2 #mid는 start
        if x == a[mid]: #만약 x 가 a리스트의 mid번째에 있는 값과 같다면
            cnt += 1
            print("반복 횟수: %d" %cnt)
            return mid
        elif x>a[mid]: #만약 x가 a리스트의 mid번째에 있는 값보다 크다면
            cnt += 1
            start = mid + 1
        else: #아니면
            cnt += 1
            end = mid - 1 #end에 min-1 저장
    cnt += 1
    print("반복 횟수: %d" %cnt)
    return -1 #-1 돌려주기
d = [10, 23, 35, 47, 51, 62, 78, 81, 97] #d리스트 정의
while(True): #무한반복
    cnt = 0
    a = int(input("데이터 입력: ")) #a에 입력받기
    if a == 0 :#만약 0이라면
        break #종료
    print(binary_search(d, a, cnt)) #결과 출력
```

```python
#(2)
def binary_recur(a, x, start, end, cnt): #binary_recur 함수 정의
    cnt += 1
    if start > end: #만약 start가 end보다 크다면
        print("반복 횟수: %d" %cnt)
        return -1 # -1 돌려주기
    mid = (start + end)//2 #mid에 (start + end)//2 한거를 저장
    if x == a[mid]: #만약 x 가 a리스트의 mid번째에 있는 값과 같다면
        print("반복 횟수: %d" %cnt)
        return mid #mid 돌려주기
    elif x > a[mid]: #만약 x가 a리스트의 mid번째에 있는 값보다 크다면
        return binary_recur(a, x, mid + 1, end, cnt) #binary_recur 함수 실행
    else: #아니면
        return binary_recur(a, x, start, mid -1, cnt) #binary_recur함수 실행
def binary_recursion(a,x): #binary_recursion 함수 정의
    cnt = 0
    return binary_recur(a, x, 0, len(a) - 1 ,cnt) #binary_recur 함수 실행
d = [10, 23, 35, 47, 51, 62, 78, 81, 97] #d리스트 생성
while(True): #무한반복
    a = int(input("데이터 입력:")) #a에 값 입력받기
    if a == 0: #만약 0이라면
        break #종료
    print(binary_recursion(d, a)) #결과 출력
```
