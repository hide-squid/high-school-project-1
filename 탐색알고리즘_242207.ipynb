```python
def search_list(a, x): # 함수 search_list선언(a = v, x = a)
    n = len(a) # n에 a리스트 길이 저장
    count = 0
    for idx in range(n): # 반복문 n-1만큼 반복
        count += 1
        if x == a[idx]: # x가 a[idx]일때
            print("탐색횟수 : %d" %count)
            return idx # idx 넘겨주기 (주소값)
    return -1 # -1 넘겨주기 (리스트에 없는 값)
v = [17, 92, 18, 32, 52, 7, 33, 35] # v 리스트 선언
while(True): #무한 반복
    a = int(input("데이터 입력: ")) # a에 입력값 저장
    print(search_list(v, a)) # idx나 -1 출력
    if a == 0:
        break
```

1. 순차 탐색 알고리즘의 정의 및 특징
 - 리스트 안에 있는 특정한 데이터를 찾기 위해 앞에서부터 데이터를 하나씩 차례대로 확인하는 방법입니다.
2. 빈칸의 (가), (나)에 들어갈 코드
 - (가) : n
 - (나) : idx
3. 33을 입력했을 때 출력 결과와 탐색 횟수
 - 탐색 과정은 0부터 7까지 반복하여 리스트 안에 있는 숫자가 33일때 까지 실행한다. 출력결과는 6회, 탐색 횟수는 7회 입니다.
