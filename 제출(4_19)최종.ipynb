<for 반복문>
1. for 변수 in range(숫자)
    - 변수값이 0~(숫자-1)까지 +1만큼 증가하여 반복됨(숫자만큼 반복이 이루어짐)
2. for 변수 in range(숫자1, 숫자2)
    - 변수값이 숫자1~(숫자2-1)까지 +1만큼 증가하여 반복됨
3. for 변수 in range(숫자1, 숫자2, 숫자3)
    - 변수값이 숫자1~(숫자2-1)까지 숫자3만큼 증가하여 반복됨




```python


limit = int(input("숫자를 입력하세여: ")) #2부터 limit 까지
print("출력결과는: ")
for num in range(2, limit + 1): #limit에 5를 넣었을때, 2부터 5까지의 수 중 소수 판별
    is_p = True #is_p의 역학은 소수인지 아닌지 구분해주기 위한 역학
    for i in range(2, int(num**0.5) + 1): #오떤 정수이 제곱근 이하의 모든 자연수로 나눠주는 상황
        if num % i == 0:
            is_p = False #즉, 소수가 아닌 경우에는 is_p 변수값을 false 바꿔준다.
            break #break가 들어있는 부모 반복문을 빠져나오는 기능(여기에서는 8번줄 for)
    print(num)
    if is_p: # is_p 변수 값이 True일때 소수 출력이 이루어짐
        print("%d@" %num, end = "")
```

while문 파악하기

```python
i=1 #변수 i에 1저장
while i<=10: #반복문 while을 1부터 10까지 반복
    print(i) # i출력
    i=i+1 #변수 i 값에 1를 더함(반복문이 끝날때 까지)
```

while 뒤에 조건:
뒤에 조건을 적을수 있다.

```python
i=1 #변수 i에 1저장
while i<=10: #반복문 while을 1부터 10까지 반복
    print(i) # i출력
    i=i+2 #변수 i 값에 2를 더함(반복문이 끝날때 까지)
```

```python
i=1 #변수 i에 1저장
while True: #반복문 while을 무한반복
    print(i) #i출력
    i=i+2 #i, i+2, i+4, i+6, i+8
    if i>10: #i가 10보다 크면 11이 나오면
        break #무한 반복을 멈춘다
```
